```python
from sklearn import datasets
iris=datasets.load_iris()
```

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
data=pd.read_csv('Iris.csv')
data.head()
```

```python
data.shape
```

```python
x=data.drop('Species', axis=1)
y=data['Species']
```

```python
from sklearn.preprocessing import LabelEncoder
Lb=LabelEncoder()
data['Species']=Lb.fit_transform(data['Species'])
data.head()
```

```python
from sklearn.cluster import KMeans
km=KMeans(n_clusters=3)
km
```

```python
y_pred=km.fit_predict(data)

data['clusters']=y_pred
data.head()
```

```python
data.clusters.unique()
```

```python
df1=data[data.clusters==0]
df2=data[data.clusters==1]
df3=data[data.clusters==2]
```

```python
plt.scatter(df1['PetalLengthCm'], df1['PetalWidthCm'], color='red')
plt.scatter(df2['PetalLengthCm'], df2['PetalWidthCm'], color='blue')
plt.scatter(df3['PetalLengthCm'], df3['PetalWidthCm'], color='black')
plt.xlabel('PetalLenghtcm')
plt.ylabel('PetalWidthCm')
```
