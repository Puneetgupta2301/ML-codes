```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
import warnings

data=pd.read_csv('sonar_dataset.csv',header=None)
```

```python
data.head()
```

```python
data.shape
```

```python
data.isna().sum()
```

```python
X=data.drop(60,axis='columns')
y=data[60]
```

```python
X
```

```python
y
```

```python
for i in data.columns:
    data[i].replace({'R':0, 'M':1}, inplace=True)
```

```python
y
```

```python
pd.get_dummies(y,drop_first=True)
data.sample(5)
```

Train_test_split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
y_train
```

```python
import tensorflow as tf
from tensorflow import keras
model=keras.Sequential([
    keras.layers.Dense(60,input_dim=60,activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(30,activation='relu'),
    keras.layers.Dropout(0.2),
    keras.layers.Dense(15,activation='relu'),
    keras.layers.Dropout(0.3),
    keras.layers.Dense(1,activation='sigmoid'),
])
model.compile(loss='binary_crossentropy',metrics=['accuracy'],optimizer='adam')
model.fit(X_train,y_train, epochs=96,batch_size=8)
```

```python
model.evaluate(X_test,y_test)
```

```python
y_pred=model.predict(X_test)
```

```python
yp=[]
for i in y_pred:
  if i>0.5:
    yp.append(1)
  else:
    yp.append(0)
```

```python
yp[:10]
```

```python
y_test[:10]
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,yp))
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(yp,y_test)
```
